```python
for i in range(5):  # 0 to 4
    print(i)
```

```python
for i in range(2, 10, 2):  # 2, 4, 6, 8
    print(i)
```

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
for i in range(6):
    if i == 3:
        continue
    print(i)
```

```python
for i in range(1, 4):
    for j in range(1, 4):
        print(i, j)
```

```python
fruits = ["apple", "banana", "cherry"]
for index, fruit in enumerate(fruits):
    print(index, fruit)
```
