```python
def greet(lang):
    messages = {
        "en": "Hello",
        "hi": "Namaste",
        "fr": "Bonjour",
    }
    return messages.get(lang, "Language not supported")

print(greet("hi"))
```

```python
cache = {}

def square(n):
    if n in cache:
        return cache[n]
    
    cache[n] = n * n
    return cache[n]

print(square(10))
print(cache)
```
