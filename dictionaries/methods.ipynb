```python
#. clear() - Removes all items from the dictionary

d = {'a': 1, 'b': 2}
d.clear()
print(d)  # Output: {}
```

```python
## 2. get(key, default=None) - Returns value for key if present, else default
d = {'a': 1, 'b': 2}
print(d.get('a'))        # Output: 1
print(d.get('c', 'N/A')) # Output: N/A
```

```python
#3. items() - Returns key-value pairs as a view object
d = {'a': 1, 'b': 2}
print(list(d.items()))  # Output: [('a', 1), ('b', 2)]
```

```python
#keys() – Returns a view of dictionary keys.

d = {'a': 1, 'b': 2}
print(list(d.keys()))  # Output: ['a', 'b']
```

```python
##pop(key, default) – Removes and returns the value for keyd = {'a': 1, 'b': 2}
print(d.pop('a'))      # Output: 1
print(d.pop('c', 0))   # Output: 0
print(d)               # Output: {'b': 2}


```

```python
##update(other_dict) – Updates dictionary with key-value pairs from another dictionary.

d = {'a': 1, 'b': 2}
d.update({'b': 3, 'c': 4})
print(d)  # Output: {'a': 1, 'b': 3, 'c': 4}
```

```python
#setdefault(key, default)
d = {'a': 1}
print(d.setdefault('b', 2))  # Output: 2
print(d)                     # Output: {'a': 1, 'b': 2}
```

```python
##copy() – Returns a shallow copy of the dictionary.

d = {'a': 1}
d_copy = d.copy()
print(d_copy)  # Output: {'a': 1}
```

```python
##popitem() – Removes and returns the last inserted key-value pair.

d = {'a': 1, 'b': 2}
print(d.popitem())  # Output: ('b', 2)
print(d)            # Output: {'a': 1}
```

```python
###fromkeys(iterable, value=None) 
keys = ['a', 'b', 'c']
d = dict.fromkeys(keys, 0)
print(d)  # Output: {'a': 0, 'b': 0, 'c': 0}
```
