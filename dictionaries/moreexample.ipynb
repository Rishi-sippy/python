```python
student = {
    "name": "Alice",
    "age": 20,
    "major": "Computer Science"
}
print(student)

```

```python
student = {"name": "Alice", "age": 20, "major": "Computer Science"}
print(student["name"])  # Output: Alice
```

```python
student["age"] = 21  # Modify existing value
student["gpa"] = 3.8  # Add new key-value pair
print(student)
# Output: {'name': 'Alice', 'age': 21, 'major': 'Computer Science', 'gpa': 3.8}
```

```python
student = {"name": "Alice", "age": 20}
print(student.get("age"))  # Output: 20
print(student.get("gpa", "Not found"))  # Output: Not found
```

```python
student = {"name": "Alice", "age": 20, "major": "Computer Science"}
for key in student:
    print(key)
# Output: name, age, major

```

```python
for value in student.values():
    print(value)
# Output: Alice, 20, Computer Science
```

```python
for key, value in student.items():
    print(f"{key}: {value}")
# Output:
# name: Alice
# age: 20
# major: Computer Science
```

```python
students = {
    "alice": {
        "age": 20,
        "major": "Computer Science",
        "gpa": 3.8
    },
    "bob": {
        "age": 21,
        "major": "Engineering",
        "gpa": 3.6
    }
}

# Accessing nested values
print(students["alice"]["major"])  # Output: Computer Science
print(students["bob"]["gpa"])  # Output: 3.6
```

```python
numbers = [1, 2, 3, 4, 5]
squared = {num: num**2 for num in numbers}
print(squared)
# Output: {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
```

```python
numbers = {1: 'one', 2: 'two', 3: 'three', 4: 'four', 5: 'five'}
even_dict = {k: v for k, v in numbers.items() if k % 2 == 0}
print(even_dict)
# Output: {2: 'two', 4: 'four'}
```

```python
numbers = range(1, 6)
label_dict = {num: "even" if num % 2 == 0 else "odd" for num in numbers}
print(label_dict)
# Output: {1: 'odd', 2: 'even', 3: 'odd', 4: 'even', 5: 'odd'}
```
