```python
# Creating a list of strings
fruits = ["apple", "banana", "cherry"]

# Creating a list with mixed data types
mixed_list = [1, "apple", 3.14, True]

# Creating an empty list
empty_list = []
```

```python
print(fruits[0])   # Output: apple
print(fruits[-1])  # Output: cherry (last element)
print(fruits[1:3]) # Output: ['banana', 'cherry'] (slice)
```

```python
fruits.append("date")        # Add item at the end
fruits.insert(1, "orange")   # Insert at position 1
fruits.remove("banana")      # Remove first occurrence by value
removed_item = fruits.pop()  # Remove and return last item
```
