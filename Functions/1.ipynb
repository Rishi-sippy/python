```python
def maths(a, b):
    sum_val = a + b
    product = a * b
    return sum_val, product

x, y = maths(5, 10)
print(x, y)
```

```python
def total(numbers):
    return sum(numbers)

print(total([10, 20, 30, 40]))
```

```python
def get_evens(nums):
    return [n for n in nums if n % 2 == 0]

print(get_evens([1,2,3,4,5,6,7,8]))
```

```python
def multiply(*nums):
    result = 1
    for n in nums:
        result *= n
    return result

print(multiply(2, 3, 4))
```

```python
def person_info(**info):
    for key, value in info.items():
        print(key, ":", value)

person_info(name="Rishi", age=25, city="Bangalore")
```

```python
pairs = [(1, 'b'), (3, 'a'), (2, 'c')]
pairs_sorted = sorted(pairs, key=lambda x: x[1])
# [(3, 'a'), (1, 'b'), (2, 'c')]
```

```python
def make_multiplier(n):
    return lambda x: x * n

doubler = make_multiplier(2)
tripler = make_multiplier(3)

print(doubler(5))  # 10
print(tripler(5))  # 15
```
