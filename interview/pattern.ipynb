```python
def print_star_pattern(n):
    for i in range(1, n + 1):
        print('*' * i)

print_star_pattern(5)
```

```python
def print_inverted_star_pattern(n):
    for i in range(n, 0, -1):
        print('*' * i)

print_inverted_star_pattern(5)
```

```python
def print_number_pyramid(n):
    for i in range(1, n + 1):
        print(' ' * (n - i) + ' '.join(str(x) for x in range(1, i + 1)))

print_number_pyramid(5)
```

```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

print(is_prime(17))  # True
```

```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=' ')
        a, b = b, a + b

fibonacci(7)
```

```python
def reverse_string(s):
    result = ''
    for char in s:
        result = char + result
    return result

print(reverse_string("hello"))  # 'olleh'
```

```python
def factorial(n):
    result = 1
    for i in range(2, n+1):
        result *= i
    return result

print(factorial(5))  # 120
```

```python
def reverse_string_name(s):
  result = ''
  for char in s:
    result = char + result
  return result
print(reverse_string_name('Rishi'))
```
