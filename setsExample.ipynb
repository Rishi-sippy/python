```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

# Union: all unique elements from both sets
print(a.union(b))  # {1, 2, 3, 4, 5, 6}
print(a | b)       # {1, 2, 3, 4, 5, 6}

# Intersection: elements common to both sets
print(a.intersection(b))  # {3, 4}
print(a & b)              # {3, 4}

# Difference: elements in a but not in b
print(a.difference(b))  # {1, 2}
print(a - b)            # {1, 2}

# Symmetric Difference: elements in either set, but not both
print(a.symmetric_difference(b))  # {1, 2, 5, 6}
print(a ^ b)                      # {1, 2, 5, 6}
```

```python
a.update(b)      # Adds elements from b into a
a.intersection_update(b)  # Updates a to keep only elements also in b
a.difference_update(b)    # Removes elements found in b from a
a.symmetric_difference_update(b)  # Updates a to elements in either set but not both
```

```python
# Set of squares of even numbers from 0 to 9
squares = {x**2 for x in range(10) if x % 2 == 0}
print(squares)  # {0, 4, 16, 36, 64}
```
